```python
import pandas as pd

klue_df = pd.read_csv("../klue/klue_preprocessing.csv")
syllaubs_df = pd.read_csv("../Syllabus/syllabus_preprocessing.csv")
```

```python
klue_df = klue_df[~klue_df["Semester"].isin(["여름학기", "겨울학기"])]
klue_df["Semester"] = klue_df["Semester"].str.replace("학기", "R", regex=False)

klue_df
```

```python
syllaubs_df.rename(columns={"학수번호": "ID", "년도": "Year", "분반": "Class", "학기": "Semester"}, inplace=True)
syllaubs_df
```

```python
print(len(klue_df))
print(len(syllaubs_df))
```

```python
if not syllaubs_df.duplicated(subset=["Year", "ID", "Semester", "Class"]).any():
    print("id + date는 유일한 키입니다.")
else:
    print("id + date 조합에 중복이 있습니다.")
```

```python
key_cols = ["Year", "ID", "Semester", "Class"]

for col in key_cols:
    klue_df[col] = klue_df[col].astype(str)
    syllaubs_df[col] = syllaubs_df[col].astype(str)
# 이제 merge 수행
merged_df = pd.merge(klue_df, syllaubs_df, on=key_cols, how="inner")
```

```python
merged_df
```

```python
merged_df.drop(columns=["특별유형_MOOC", "성적평가_상대평가", "특별유형_Flipped Class(거꾸로수업)"], inplace=True)
```

```python
merged_df.drop(columns=["특별유형_유연학기(집중수업)", "Unnamed: 0_y"], inplace=True)
```

```python
merged_df.drop(columns=["특별유형_네모클래스", "특별유형_튜토리얼", "Unnamed: 0_x"], inplace=True)
```

```python
merged_df.to_csv("merged_preprocessing.csv", encoding="utf-8", index_label=False)
```
