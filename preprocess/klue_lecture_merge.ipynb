```python
import numpy as np
import pandas as pd
import openpyxl

prefix = "klue_lectures_desc_"
postfix = ".xlsx"

df_list = []

for i in range(4):
    s = prefix + str(i + 1) + postfix
    wb = openpyxl.load_workbook(s)
    ws = wb["Sheet"]
    data = ws.values
    columns = next(data)[0:]
    df = pd.DataFrame(data, columns=columns)
    df_list.append(df)

klue_df = pd.concat(df_list)
lecture_df = pd.read_csv("final.csv")
```

```python
klue_df
klue_df = klue_df[~klue_df["Semester"].isin(["여름학기", "겨울학기"])]
klue_df["Semester"] = klue_df["Semester"].str.replace("학기", "R", regex=False)

klue_df
```

```python
lecture_df.rename(columns={"학수번호": "ID", "년도": "Year", "분반": "Class", "학기": "Semester"}, inplace=True)
lecture_df
```

```python
print(len(klue_df))
print(len(lecture_df))
```

```python
if not lecture_df.duplicated(subset=["Year", "ID", "Semester", "Class"]).any():
    print("id + date는 유일한 키입니다.")
else:
    print("id + date 조합에 중복이 있습니다.")
```

```python
key_cols = ["Year", "ID", "Semester", "Class"]

for col in key_cols:
    klue_df[col] = klue_df[col].astype(str)
    lecture_df[col] = lecture_df[col].astype(str)
# 이제 merge 수행
merged_df = pd.merge(klue_df, lecture_df, on=key_cols, how="inner")
```
