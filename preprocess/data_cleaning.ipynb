```python
import pandas as pd

df = pd.read_csv("merged.csv")
```

```python
df
```

```python
df["recc_rate"] = (
    df["recc_rate"]
    .str.rstrip('%')        # '%' 기호 제거
    .astype(float)          # 문자열 → float 변환
    .div(100)               # 100으로 나눠서 0~1 사이 백분율로
)
```

```python
df
```

```python
df.drop(columns=["dept_cd", "detailed_info"], inplace=True)
```

```python
dummies = pd.get_dummies(df["Attd_rate"], prefix="출석")
df = pd.concat([df, dummies], axis=1)
```

```python
df.drop(columns=["Attd_rate"], inplace=True)
```

```python
df.to_csv("final_data.csv", index=False, encoding="cp949")
```
