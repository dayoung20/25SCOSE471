#### 2020년, 2021 - 2023년의 teaching ability

```python
import pandas as pd
import matplotlib.pyplot as plt

# 2020년의 수업유형_대면은 모두 False로 처리
df.loc[df['Year'] == 2020, '수업유형_대면'] = False

# 그룹 분리
group_2020 = df[df['Year'] == 2020]
group_2021_2023 = df[df['Year'].isin([2021, 2022, 2023])]

# 각 그룹의 teaching ability 평균 계산
mean_2020 = group_2020['teaching ability'].mean()
mean_2021_2023 = group_2021_2023['teaching ability'].mean()

# 시리즈로 묶어서 시각화
mean_series = pd.Series({
    '2020': mean_2020,
    '2021~2023': mean_2021_2023
})

# 시각화
plt.figure(figsize=(6, 4))
mean_series.plot(kind='bar', color=['steelblue', 'coral'])
plt.title('Teaching Ability 평균 비교: 2020 vs 2021~2023')
plt.ylabel('Teaching Ability 평균')
plt.xlabel('연도 구간')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
#### Kmeasn 클러스터링

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# 분석에 필요한 변수 선택
features = ['teaching ability', 'Year', '수업유형_대면']  # 이진형 변수 포함
data = df[features].dropna()

# 수업유형_대면을 0/1로 변환 (이미 bool이면 그대로 사용 가능)
data['수업유형_대면'] = data['수업유형_대면'].astype(int)

# 정규화
scaler = StandardScaler()
scaled = scaler.fit_transform(data)

# KMeans 적용 (클러스터 2개 가정: 코로나 영향 전후)
kmeans = KMeans(n_clusters=2, random_state=42)
data['cluster'] = kmeans.fit_predict(scaled)

# 시각화
sns.boxplot(x='cluster', y='teaching ability', data=data)
plt.title("Teaching Ability의 클러스터별 분포")
plt.show()

# 연도별 클러스터 분포 확인
pd.crosstab(data['Year'], data['cluster'])
```

```python
import matplotlib.pyplot as plt

year_dist = pd.crosstab(data['Year'], data['cluster'])

# 막대 그래프
year_dist.plot(kind='bar', figsize=(8, 5))
plt.title("Cluster별 연도 분포")
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend(title="Cluster")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
### 클러스터링 evaluation

from sklearn.metrics import silhouette_score

score = silhouette_score(scaled, data['cluster'])
print(f"Silhouette Score: {score:.3f}")
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled)

plt.figure(figsize=(6, 5))
sns.scatterplot(x=pca_data[:, 0], y=pca_data[:, 1], hue=data['cluster'], palette='Set1')
plt.title("PCA로 본 클러스터 분포")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.tight_layout()
plt.show()
```

`cluster 0 `
- 2021년 이후부터 급격히 많아지고 2022, 2023년에는 거의 모든 데이터가 클러스터 0에 속함.
- 코로나 이후 시대의 수업 유형과 teaching ability 특성을 반영하는 집단

`cluster 1` :
- 2017~2020년에 많이 분포하고 2022-2023년에는 거의 사라짐.
- 코로나 이전의 수업 환경을 더 많이 반영하고 있는 집단

#### 코로나 이전과 이후를 그룹화하여 영향을 미친 요소 분석

```python
import pandas as pd

# 수치형 변수를 위한 전처리
df_numeric = df.copy()
df_numeric = df_numeric.apply(pd.to_numeric, errors='coerce')

# 코로나 이전과 이후로 나누기
pre_covid = df_numeric[df_numeric['Year'] <= 2019]
post_covid = df_numeric[df_numeric['Year'] >= 2020]

# 상관관계 계산 (teaching ability 제외)
pre_corr = pre_covid.corr()['teaching ability'].drop('teaching ability').dropna()
post_corr = post_covid.corr()['teaching ability'].drop('teaching ability').dropna()

# 상관관계 절댓값 기준 상위 10개 추출
top10_pre = pre_corr.abs().sort_values(ascending=False).head(10)
top10_post = post_corr.abs().sort_values(ascending=False).head(10)

# 결과 출력
print("코로나 이전 (≤2019) - teaching ability 상관 높은 변수 Top 10:")
print(pre_corr.loc[top10_pre.index].sort_values(ascending=False), "\n")

print("코로나 이후 (≥2020) - teaching ability 상관 높은 변수 Top 10:")
print(post_corr.loc[top10_post.index].sort_values(ascending=False))
```

#### 결과 분석

**코로나 이전**
- attendance rate(0.978)로 가장 높은 상관관계를 가지며, 다른 요소들은 attendance rate에 비해 아주 약한 상관관계를 가짐
- 특히 출석확인자율화(-0.450)가 강한 음의 상관관계를 가지는 것으로 보아 출석확인 자율화가 도입된 경우에 teaching ability가 낮아지는 강한 관계를 가짐

**코로나 이후**
- attendance rate가 0.978 -> 0.332로 크게 감소하였음
- 비대면 수업이 많아지면서 출석이 평가에 반영되는 비율이 줄었거나 학생들의 교수 평가 방식에 변화가 생긴 것을 의미
- 코로나 이전에는 attendance rate 이외의 요소가 teaching ability에 거의 영향을 미치지 않았다면 코로나 이후에는 achievement level,course evaluation과 같은 요소의 영향이 커짐

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# teaching ability 및 수치형 변수만 필터링
numeric_df = df.select_dtypes(include=['number', 'bool']).copy()
numeric_df = numeric_df.dropna(subset=['teaching ability', 'Year'])

# 수업유형_대면을 정수형으로 변환
if '수업유형_대면' in numeric_df.columns:
    numeric_df['수업유형_대면'] = numeric_df['수업유형_대면'].astype(int)

# 코로나 이전/이후로 나누기
before_covid = numeric_df[numeric_df['Year'] <= 2019].copy()
after_covid = numeric_df[numeric_df['Year'] >= 2020].copy()

# 클러스터링 함수 정의
def cluster_and_plot(data, title):
    scaler = StandardScaler()
    scaled = scaler.fit_transform(data)

    kmeans = KMeans(n_clusters=2, random_state=42)
    data['cluster'] = kmeans.fit_predict(scaled)

    # 박스플롯 시각화
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='cluster', y='teaching ability', data=data)
    plt.title(f"{title} - Teaching Ability by Cluster")
    plt.tight_layout()
    plt.show()

    return data

# 클러스터링 및 시각화
clustered_before = cluster_and_plot(before_covid[['teaching ability', 'Year', '수업유형_대면']], "Before COVID-19")

clustered_after = cluster_and_plot(after_covid[['teaching ability', 'Year', '수업유형_대면']], "After COVID-19")
```

#### 코로나 이전과 이후의 클러스터링 분석 결과

**코로나 이전**

`cluster 0` : teaching ability의 평균이 높고, 전반적으로 우수한 평가를 받는 수업들이 속함

`cluster 1` : teaching ability가 낮은 수업들이 모여있음

클러스터 간 분리가 명확하며, 두 그룹 간의 차이가 큼 -> teaching ability에 의해 평가가 뚜렷하게 나뉨

**코로나 이후**

두 클러스터 모두 teaching ability의 분포가 유사하며 큰 차이를 보이지 않음

코로나 이후 비대면 강의가 일반화되면서 학생들이 교강사의 강의력 차이를 느끼기 어려웠을 가능성이 있음

실제 강의 내용보다 플랫폼, 네트워크 환경, 과제 중심 평가 등이 영향을 주었을 수 있음
